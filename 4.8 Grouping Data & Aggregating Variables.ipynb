### Ex 6. Creating a spender flag

```python
# 6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by
#    looking at the prices of the items people are buying. Create a spending flag for each user based on the 
#    average price across all their orders using the following criteria:

# - If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
# - If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a 
#   “High spender.”
```

```python
# Create a column for the average price across a user's orders

ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check whether the column is correct

ords_prods_merge.head(100)
```

```python
# I am looking at all entries for the user with the user_id 1 to check whether the mean is correct

ords_prods_merge.loc[ords_prods_merge['user_id'] == 1]
```

```python
# Create the Spender Flag

ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spender_flag'] = 'High Spender'
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spender_flag'] = 'Low Spender'
```

```python
# Print the frequency of your new “spender_flag” column using the value_counts() function

ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
# Checking whether the flag is correct

ords_prods_merge[['user_id', 'spender_flag', 'average_price']].head(300) 
```

### Ex. 7 Creating a flag for order frequency

```python
# 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to 
#    buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. 
#    Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median
#    in the “days_since_prior_order” column. The criteria for the flag should be as follows:

# - If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a 
#   “Non-frequent customer.”
# - If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a 
#   “Regular customer.”
# - If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”
```

```python
# Create a column for the median in "days_since_prior_order" column

ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Print the frequency of your new “median_days_since_prior_order” column using the value_counts() function

ords_prods_merge['median_days_since_prior_order'].value_counts(dropna = False)
```

```python
# Checking if the NaN frequency is correct. 
# The median_days_since_prior_order value can only be NaN if only 1 order has been placed by the customer.
# There are five entries for customers who placed only one order, which checks out with the NaN count above. 

ords_prods_merge.loc[ords_prods_merge['max_order'] == 1]
```

```python
# Creating the frequency_flag column 

ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10) & (ords_prods_merge['median_days_since_prior_order'] <= 20) , 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Print the frequency of your new “frequency_flag” column using the value_counts() function

ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# We could address the NaN issue by changing the condition for non-frequent customers to: 
# If the median of “days_since_prior_order” is higher than 20 or Nan, then the customer should be labeled a 
# “Non-frequent customer.”
```

```python
# Checking whether the frequency flag is correct

ords_prods_merge[['user_id', 'frequency_flag', 'median_days_since_prior_order']].head(100) 

# Looks correct
```
